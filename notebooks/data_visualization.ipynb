```python
# Importing libraries

import matplotlib.pyplot as plt
import pandas as pd
import os
```

```python
# Importing dataset

hotel_dataset = pd.read_csv('../data/hotel_bookings.csv')
```

```python
# Summary of number of adults

counts = hotel_dataset['no_of_adults'].value_counts().sort_index()

plt.bar(counts.index, counts.values, color='lightblue')

plt.title('Number of Adults per Booking')
plt.xlabel('Adults')
plt.ylabel('Frequency')
plt.xticks(counts.index)
plt.grid(axis='y', linestyle='--', alpha=0.3)

# Keep only left and bottom spines
ax = plt.gca()
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)

plt.savefig("../plots/no_of_adults.png", dpi=300, bbox_inches='tight')

plt.show()
```

```python
# Summary of number of children

counts = hotel_dataset['no_of_children'].value_counts().sort_index()

plt.bar(counts.index, counts.values, color='lightblue')

plt.title('Number of Children per Booking')
plt.xlabel('Children')
plt.ylabel('Frequency')
plt.xticks(counts.index)
plt.xlim(counts.index.min() - 0.5, counts.index.max() + 0.5)
plt.grid(axis='y', linestyle='--', alpha=0.3)

ax = plt.gca()
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)

plt.savefig("../plots/no_of_children.png", dpi=300, bbox_inches='tight')
plt.show()
```

```python
# Summary of weekend nights

counts = hotel_dataset['no_of_weekend_nights'].value_counts().sort_index()

plt.bar(counts.index, counts.values, color='lightblue')

plt.title('Number of Weekend Nights per Booking')
plt.xlabel('Weekend Nights')
plt.ylabel('Frequency')
plt.xticks(counts.index)
plt.grid(axis='y', linestyle='--', alpha=0.3)

# Remove top and right spines
ax = plt.gca()
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)

plt.savefig("../plots/no_of_weekend_nights.png", dpi=300, bbox_inches='tight')

plt.show()
```

```python
# Summary of week nights

counts = hotel_dataset['no_of_week_nights'].value_counts().sort_index()

fig, ax = plt.subplots()

ax.bar(counts.index.astype(str), counts.values, color='lightblue')

# Labels and title
ax.set_title('Number of Week Nights per Booking')
ax.set_xlabel('Week Nights')
ax.set_ylabel('Frequency')
ax.grid(axis='y', linestyle='--', alpha=0.3)

# Remove top and right spines
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)

# Sparse x-tick labels
xticks = counts.index
sparse_labels = [str(x) if x % 2 == 0 else '' for x in xticks]
ax.set_xticks(xticks)
ax.set_xticklabels(sparse_labels)

plt.savefig("../plots/no_of_week_nights.png", dpi=300, bbox_inches='tight')
plt.show()
```

```python
# Summary of required parking space

counts = hotel_dataset['required_car_parking_space'].value_counts().sort_index()

fig, ax = plt.subplots()

ax.bar(counts.index.astype(str), counts.values, color='lightblue')

# Labels and title
ax.set_title('Required Parking Space')
ax.set_xlabel('Parking Space')
ax.set_ylabel('Frequency')
ax.grid(axis='y', linestyle='--', alpha=0.3)

# Remove top and right spines
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)

plt.savefig("../plots/required_car_parking_space.png", dpi=300, bbox_inches='tight')
plt.show()
```

```python
# Summary of time between booking and arrival date

lead_times = hotel_dataset['lead_time']

fig, ax = plt.subplots()

# Histogram for continuous lead time
ax.hist(lead_times, bins=15, color='lightblue')

# Labels and title
ax.set_title('Lead Days')
ax.set_xlabel('Lead Days')
ax.set_ylabel('Frequency')
ax.set_xlim(-1, 443)
ax.grid(axis='y', linestyle='--', alpha=0.3)

# Clean look
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)

plt.savefig("../plots/lead_time_hist.png", dpi=300, bbox_inches='tight')
plt.show()
```

```python
# Summary of arrival month

counts = hotel_dataset['arrival_month'].value_counts().sort_index()

fig, ax = plt.subplots()

# Bar chart for discrete months
ax.bar(counts.index.astype(str), counts.values, color='lightblue')

# Labels and title
ax.set_title('Arrival Month')
ax.set_xlabel('Arrival Month')
ax.set_ylabel('Frequency')
ax.grid(axis='y', linestyle='--', alpha=0.3)

# Clean spines
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)

plt.savefig("../plots/arrival_month_bar.png", dpi=300, bbox_inches='tight')
plt.show()
```

```python
# Summary of arrival day

counts = hotel_dataset['arrival_date'].value_counts().sort_index()

fig, ax = plt.subplots()

# Use integer x-values for alignment
x = counts.index
y = counts.values

# Bar chart with integer x (not strings)
ax.bar(x, y, color='lightblue')

# Labels and title
ax.set_title('Arrival Day')
ax.set_xlabel('Arrival Day')
ax.set_ylabel('Frequency')
ax.grid(axis='y', linestyle='--', alpha=0.3)

# Clean spines
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)

# Set x-ticks and sparse labels (every 3rd)
ax.set_xticks(x)
sparse_labels = [str(val) if val % 3 == 0 else '' for val in x]
ax.set_xticklabels(sparse_labels)

plt.savefig("../plots/arrival_date_bar.png", dpi=300, bbox_inches='tight')
plt.show()
```

```python
# Summary of repeated guests

counts = hotel_dataset['repeated_guest'].value_counts().sort_index()

fig, ax = plt.subplots()

# Bar chart for binary variable
ax.bar(counts.index, counts.values, color='lightblue')

# Labels and title
ax.set_title('Repeated Guest')
ax.set_xlabel('Repeated Guest')
ax.set_ylabel('Frequency')
ax.grid(axis='y', linestyle='--', alpha=0.3)

# Clean spines
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)

# x-ticks: 0 and 1
ax.set_xticks(counts.index)
ax.set_xticklabels([str(x) for x in counts.index])

plt.savefig("../plots/repeated_guest_bar.png", dpi=300, bbox_inches='tight')
plt.show()
```

```python
# Summary of average prices

prices = hotel_dataset['avg_price_per_room']

fig, ax = plt.subplots()

# Histogram for continuous variable
ax.hist(prices, bins=30, color='lightblue')

# Labels and title
ax.set_title('Average Price per Room')
ax.set_xlabel('Average Price')
ax.set_ylabel('Frequency')
ax.set_xlim(0, 365)
ax.grid(axis='y', linestyle='--', alpha=0.3)

# Clean spines
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)

plt.savefig("../plots/avg_price_per_room_hist.png", dpi=300, bbox_inches='tight')
plt.show()

```
