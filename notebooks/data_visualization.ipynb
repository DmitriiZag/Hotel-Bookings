```python
# Importing libraries

import matplotlib.pyplot as plt
import pandas as pd
import os
```

```python
# Importing dataset

hotel_dataset = pd.read_csv('../data/hotel_bookings.csv')
```

```python
# Summary of number of adults

counts = hotel_dataset['no_of_adults'].value_counts().sort_index()

plt.bar(counts.index, counts.values, color='lightblue')

plt.title('Number of Adults per Booking')
plt.xlabel('Adults')
plt.ylabel('Frequency')
plt.xticks(counts.index)
plt.grid(axis='y', linestyle='--', alpha=0.3)

# Keep only left and bottom spines
ax = plt.gca()
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)

plt.show()
```

```python
counts = hotel_dataset['no_of_children'].value_counts().sort_index()

plt.bar(counts.index, counts.values, color='lightblue')

plt.title('Number of Children per Booking')
plt.xlabel('Children')
plt.ylabel('Frequency')
plt.xticks(counts.index)
plt.xlim(counts.index.min() - 0.5, counts.index.max() + 0.5)
plt.grid(axis='y', linestyle='--', alpha=0.3)

ax = plt.gca()
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)

plt.savefig("../plots/no_of_children.png", dpi=300, bbox_inches='tight')
plt.show()
```

```python
counts = hotel_dataset['no_of_weekend_nights'].value_counts().sort_index()

plt.bar(counts.index, counts.values, color='lightblue')

plt.title('Number of Weekend Nights')
plt.xlabel('Weekend Nights')
plt.ylabel('Frequency')
plt.xticks(counts.index)
plt.grid(axis='y', linestyle='--', alpha=0.3)

# Remove top and right spines
ax = plt.gca()
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)

plt.savefig("../plots/no_of_weekend_nights.png", dpi=300, bbox_inches='tight')

plt.show()
```
